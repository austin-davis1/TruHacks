```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def generate_random_data():
    # Generate random data for risk of outbreak, cases, current hospital capacity, and current risk
    num_days = 30
    dates = pd.date_range(start='2024-01-01', periods=num_days)
    risk_of_outbreak = np.random.randint(0, 100, size=num_days)
    cases = np.random.randint(0, 50, size=num_days)
    current_hospital_capacity = np.random.randint(50, 200, size=num_days)
    current_risk = np.random.rand(num_days) * 0.5 + 0.3  # Random values between 0.3 and 0.8 for current risk

    # Create DataFrame
    data = {
        'Date': dates,
        'Risk of Outbreak': risk_of_outbreak,
        'Cases': cases,
        'Current Hospital Capacity': current_hospital_capacity,
        'Current Risk': current_risk
    }
    df = pd.DataFrame(data)
    return df

def plot_data(df):
    # Plotting
    plt.figure(figsize=(14, 8))

    # Risk of Outbreak
    plt.subplot(2, 2, 1)
    plt.plot(df['Date'], df['Risk of Outbreak'], marker='o', color='r')
    plt.title('Risk of Outbreak')
    plt.xlabel('Date')
    plt.ylabel('Risk')

    # Bar chart for Risk of Outbreak
    plt.subplot(2, 2, 2)
    df.plot(kind='bar', x='Date', y='Risk of Outbreak', color='r', ax=plt.gca())
    plt.title('Risk of Outbreak (Bar Chart)')
    plt.xlabel('Date')
    plt.ylabel('Risk')

    # Cases
    plt.subplot(2, 2, 3)
    plt.plot(df['Date'], df['Cases'], marker='o', color='g')
    plt.title('Cases')
    plt.xlabel('Date')
    plt.ylabel('Number of Cases')

    # Current Hospital Capacity
    plt.subplot(2, 2, 4)
    plt.plot(df['Date'], df['Current Hospital Capacity'], marker='o', color='b')
    plt.title('Current Hospital Capacity')
    plt.xlabel('Date')
    plt.ylabel('Capacity')

    plt.tight_layout()
    plt.show()

# Generate and plot random data
df = generate_random_data()
plot_data(df)
```
